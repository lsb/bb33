```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
bb33 = np.genfromtxt('bb33-abc-12m-a.txt')
```

```python
bb33
```

```python
plt.plot(np.log1p(np.arange(len(bb33))), np.log1p(bb33))
```

```python
plt.plot((np.arange(len(bb33))), (bb33))
```

```python
plt.figure(figsize=(40,40))
plt.plot(np.log1p(np.arange(len(bb33)-1)), (bb33[1:] - bb33[0:-1]))
```

```python
bb33conv = np.convolve((bb33[1:] - bb33[0:-1]), np.ones(100)/100, mode="valid")
```

```python
bb33conv
```

```python
plt.figure(figsize=(10,10))
plt.plot((np.arange(len(bb33conv))), bb33conv)
```
